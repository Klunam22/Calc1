### Problem 4

Let 
$$h(x) = \begin{cases} 
          \frac{\sin(x)}{x} & x < 0 \\
          \cos(x^2) & x \geq 0 
       \end{cases}
    $$
    
Explore behaviour of $h(x)$ near $x=0$ similarly to problem 3.

```python
# Creating a graph of the function
def h(x):
    if x < 0:
        return np.sin(x) / x  # For x < 0
    else:
        return np.cos(x**2)  # For x >= 0


h = np.vectorize(h)

# Generate sequences of values approaching x = 0 from both sides
L = np.arange(1, 15, 1)
base_seq = 0.7**L

xRight = 0 + base_seq  # Approaching from the right (x >= 0)
xLeft = 0 - base_seq   # Approaching from the left (x < 0)

h_xRight = h(xRight)
h_xLeft = h(xLeft)

fig = plt.figure()
plt.plot(xLeft, h_xLeft, marker="o", color='black', label='x approaching 0 from the left')
plt.plot(xRight, h_xRight, marker="o", color='blue', label='x approaching 0 from the right')


plt.xlabel('x')
plt.ylabel('h(x)')
plt.title('Behavior of h(x) as x approaches 0')
plt.axvline(x=0, color='gray', linestyle='--', linewidth=0.7)
plt.legend()
plt.grid(True)
fig.show()

#------------------------------------------------------------------------------------#



# Calculating left and right limits

h_sympy = sp.Piecewise(
    (sp.sin(x) / x, x < 0),
    (sp.cos(x**2), x > 0)
)

left_limit = sp.N(sp.limit(h_sympy, x, 0, '-'))
right_limit = sp.N(sp.limit(h_sympy, x, 0, '+'))

print("Left Limit:", left_limit)
print("Right Limit:", right_limit)
print("The limit is continous as x approaches 0")

```

### Problem 6  (I.V.T.)

Let $f(x)=x^3+x^2-2x-2$

**a.** Evaluate $f(1)$ and $f(2)$. Looking at these values, can we conclude that there is a zero of $f(x)$ between $x=1$ and $x=2$? Why?

```python
def f(x):
    return x**3 + x**2 - 2*x - 2

#  f(1) and f(2)
x1 = 1
x2 = 2
f_x1 = f(x1)
f_x2 = f(x2)

print(f"f(1) = {f_x1}")
print(f"f(2) = {f_x2}")

# Check for sign change
if f_x1 * f_x2 < 0:
    print("There is a zero of f(x) between x=1 and x=2 by the IVT.")
else:
    print("There is no zero of f(x) between x=1 and x=2.")

#  function on the interval [1, 2]
x = np.linspace(1, 2, 500)
y = f(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label="f(x) = x^3 + x^2 - 2x - 2", color="blue")
plt.axhline(0, color="red", linestyle="--", label="y = 0")
plt.scatter([x1, x2], [f_x1, f_x2], color="green", zorder=5, label="f(1) and f(2)")

plt.title("Graph of f(x) with interval [1, 2]")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```

**Answer:** The IVT states that f(x) is continous if one value is positive and one negative then there exists one c so  f(c) = 0 

**b.** To narrow down the location of zero. Evaluate $f(1.5)$. Where is the zero now: [1,1.5] or [1.5,2]?

```python
# function
def f(x):
    return x**3 + x**2 - 2*x - 2

# Evaluate f(1) and f(2)
x1, x2 = 1, 2
f1 = f(x1)
f2 = f(x2)

print(f"f(1) = {f1}")
print(f"f(2) = {f2}")

if f1 * f2 < 0:
    print("There is a zero between x=1 and x=2 by the Intermediate Value Theorem.")
else:
    print("No zero exists between x=1 and x=2.")

# Narrow down the location of zero by evaluating f(1.5)
x_mid = 1.5
f_mid = f(x_mid)

print(f"f(1.5) = {f_mid}")

if f1 * f_mid < 0:
    print("The zero is in the interval [1, 1.5].")
elif f_mid * f2 < 0:
    print("The zero is in the interval [1.5, 2].")
else:
    print("The zero is not in the narrowed interval.")
```

**Answer:** By evaluating f(1.5) into the function f(x) we come with the conlusion that f(1.5)>0  and it lies on the interval [1,1.5] since thats when in transitions from negative to positive.
